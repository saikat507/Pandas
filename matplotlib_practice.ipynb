```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
pizza_size=np.arange(8,13)
pizza_size
```

```python
pizza_price=np.arange(300,550,50)
```

```python
pizza_price
```

```python
plt.scatter(pizza_size,pizza_price)
plt.show()
```

```python
plt.scatter(pizza_size,pizza_price,color="red")
plt.title("Pizza size vs it's price")
plt.xlabel("pizza size")
plt.ylabel("pizza price")
```

```python
pizza_size=np.arange(8,13)
pizza_price=np.arange(300,550,50)
plt.plot(pizza_size,pizza_price,color="red",marker=".",linestyle='--',linewidth=5)
```

```python
blood_group=['O+','B+','A+','AB+']
blood_group
```

```python
person=[25,7,30,20]
person
```

```python
plt.bar(blood_group,person,color=['red','green','yellow','blue'])
plt.scatter(blood_group,person)
plt.title("Blood group camping")
plt.xlabel("Blood_group")
plt.ylabel("Person")
plt.show()
```
